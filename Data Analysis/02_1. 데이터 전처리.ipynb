### 파생변수 만들기
- 파생변수는 기존의 변수들을 가지고 새로운 변수를 만들 수 있고, 기존의 컬럼 값을 가지고도 새로운 변수를 만들 수 있다.
- 파생변수는 새로운 변수지만 기존의 변수들의 관계를 가지고 만든다.

```python
#파생변수 - 기존의 변수들/컬럼 값들을 가지고 만든 새로운 변수

#.assign(만들변수의이름 = 식)
#셀에서 실행 후에 결과는 나타내주지만 실제 원 데이터에는 반영이 되지 않음
exam.assign(total_3 = (exam['math'] + exam['english'] + exam['science']) / 3,
           sum_3 = exam['math'] + exam['english'] + exam['science'])
```

```python
#원 데이터에 반영시키는 방법
#방법 1: 새로운 변수에 저장
exam_1 = exam.assign(total_3 = (exam['math'] + exam['english'] + exam['science']) / 3,
           sum_3 = exam['math'] + exam['english'] + exam['science'])
```

```python
#방법 2: inplace = True
exam.assign(total_3 = (exam['math'] + exam['english'] + exam['science']) / 3,
           sum_3 = exam['math'] + exam['english'] + exam['science'],
           inplace = True)
```

```python
import numpy as np
```

```python
# 과학 점수가 50보다 큰 값을 가진 데이터의 행은?
np.where(exam['science']>50)
```

```python
# 조건에 따라 파생변수 생성하기
# 원 데이터에 반영되지 않음

# 과학 점수가 50보다 큰 값을 가진 데이터에 True -> good, False -> bad로 파생변수 생성
exam.assign(test = np.where(exam['science']>50,'good','bad'))
```

```python
# 여러가지 과목의 점수를 합산한 값을 파생변수로 만들기
exam['tot'] = exam['english'] +exam['math'] + exam['science']
exam
```

```python
#lambda를 사용하여 파생변수 만들기
# 원데이터에 반영 x
exam.assign(new_tot = lambda x : x['math']+x['english']+x['science'])
```

```python
# groupby, agg
# 액셀의 피벗 개념과 비슷함
# 피벗: 전체 데이터에서 원하는 컬럼만 요약통계를 살펴보는 것
# groupby: 집단을 만드는 함수
# agg: 통계치를 사용할 수 있는 함수

#반 별 영어 성적의 평균 구하기
exam.groupby('nclass').agg(mean_english = ('english', 'mean')) #파생변수이름, 사용할 컬럼(변수), 원하는 요약통계치
```

```python
#반 별 영어 성적의 중앙값 구하기
exam.groupby('nclass').agg(median_english = ('english','median'))
```

```python
#여러 개를 요약통계치 만들어 보기

exam.groupby('nclass').agg(mean_math = ('math','mean'),
                         sum_math = ('math','sum'),
                         max_math = ('math','max'),
                         n = ('nclass','count')) # 데이터 개수  
#std 표준편차도 가능
```

```python
# 파생변수사용하여 새로운 파생변수 생성
exam['mean_3'] = exam['tot']/3
exam
```

```python
# merge (데이터 병합)
#데이터를 합칠 땐 공통된 기준을 갖추어 병합해야 한다.


a = pd.DataFrame({'name':['김철수','김영희','홍길동'],
                 '영어 점수':[10,20,30]})
b = pd.DataFrame({'name':['김철수','김영희','홍길길'],
                 '수학 점수':[10,20,40]})
display(a)
display(b)
```

```python
# 공통된 기준으로 병합(name)
total = pd.merge(a, b, how='inner') 
total
```

```python
#concat 
#덩어리 + 덩어리로 붙는 형태
pd.concat([a,b]) #axis 축 디폴트는 0
```

```python
pd.concat([a,b],axis=1) #1로 바꿈 
```

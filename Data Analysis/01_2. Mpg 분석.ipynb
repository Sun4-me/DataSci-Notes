```python
import pandas as pd
import numpy as np
mpg = pd.read_csv('csv/mpg.csv')
mpg
```

```python
#Q1: mpg 데이터 중 5번째로 데이터가 많은 제조사는?

#code
mpg_mf = mpg['manufacturer'].value_counts().to_frame() #manufacturer을 추출하여 각 차종의 수를 데이터프레임형태로 mpg_mf에 새롭게 저장
mpg_mf

# audi가 5번째로 데이터가 많은 제조사이다.
```

```python
#Q2: mpg데이터의 고속도로 연비의 평균은?

#code
sum(mpg['hwy']/len(mpg)) #hwy값을 모두 더하여 mpg데이터의 길이(전체 수)로 나누어 평균을 구함

#mpg데이터의 고속도로 연비의 평균은 23.44이다.
```

```python
#Q3: volkswagen의 고속도로 연비의 평균은?

#code
mpg_v = mpg[mpg['manufacturer'] == 'ford'] #ford의 데이터만 추출하여 새롭게 저장
sum(mpg_v['hwy']/len(mpg_v)) #ford의 고속도로 연비의 평균을 구함

#ford의 고속도로 연비의 평균은 19.36이다.
```

```python
#Q4: ford의 모델 중 가장 수가 많은 모델은?

#code
mpg_ford = mpg[mpg['manufacturer'] == 'ford'] #ford의 데이터만 추출하여 새롭게 저장
mpg_ford_models = mpg_ford['model'].value_counts().to_frame() #각 모델 수를 계산하여 데이터프레임 형태로 새롭게 저장
mpg_ford_models

#ford 모델 중 가장 수가 많은 모델은 mustang이다.
```

```python
#Q5: midsize 모델 중 가장 많이 제조된 모델은?

#코드 작성
mpg_c = mpg[mpg['category'] == 'midsize'] #midsize로 제조된 모델들만 추출하여 새롭게 저장
mpg_c_models = mpg_c['model'].value_counts().to_frame() #각 모델 수를 계산하여 데이터프레임 형태로 새롭게 저장
mpg_c_models

# midsize 모델 중 가장 많이 제조된 모델은 sonata이다.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

 # mpg 데이터 시각화

```python
mpg = pd.read_csv('csv/mpg.csv')
mpg
```

```python
# manufacturer groupby
mpg_grouped = mpg.groupby('manufacturer').mean().reset_index()
mpg_grouped
```

```python
# 'Average Hwy by Manufacturer' 막대 그래프
plt.figure(figsize=(12,6))
sns.barplot(data=mpg_grouped, x='manufacturer', y='hwy')
plt.xticks(rotation=45)
plt.title('Average Hwy by Manufacturer')
plt.show()
```

```python
# 'Hwy by Year and Number of Cylinders' 박스 플랏
```

```python
sns.boxplot(data=mpg, x='year', y='hwy', hue='cyl')
plt.title('Hwy by Year and Number of Cylinders')
plt.show()
```

```python
# 'Hwy by Engine Displacement and Number of Cylinders' 스캐털 플랏
sns.scatterplot(data=mpg, x='displ', y='hwy', hue='cyl')
plt.title('Hwy by Engine Displacement and Number of Cylinders')
plt.show()
```

```python
# 'Average Cty by Manufacturer' 라인 그래프
sns.lineplot(data=mpg_grouped, x='manufacturer', y='cty')
plt.xticks(rotation=45)
plt.title('Average Cty by Manufacturer')
plt.show()
```

 # titanic 데이터 시각화

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
pclass_df = titanic.groupby(['pclass', 'survived']).size().reset_index(name='Count')
pclass_df
```

```python
# 'Survival count by Pclass' 막대 그래프
sns.barplot(x='pclass', y='Count', hue='survived', data=pclass_df)
plt.title('Survival count by Pclass')
plt.show()
```

```python
# 'Survival count by Pclass' 히트맵
pclass_pivot = pclass_df.pivot('pclass', 'survived', 'Count')
sns.heatmap(pclass_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Survival count by Pclass')
plt.show()
```

```python
# 'Age distribution by Pclass and Survival' 박스 플랏
sns.boxplot(x='pclass', y='age', hue='survived', data=titanic)
plt.title('Age distribution by Pclass and Survival')
plt.show()
```

```python
# 'Survival by Age and Fare' 스캐터 플랏
sns.scatterplot(x='fare', y='age', hue='survived', data=titanic)
plt.title('Survival by Age and Fare')
plt.show()
```

```python
# 히스토그램
sns.histplot(data=titanic, x="age")
plt.title('age')
plt.show()
```

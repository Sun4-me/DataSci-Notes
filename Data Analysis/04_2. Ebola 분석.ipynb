```python
import pandas as pd
```

```python
ebola = pd.read_csv('csv/country_timeseries.csv')
ebola
```

```python
# 원본 데이터 시각화
ebola['Cases_Liberia'].plot()
```

```python
# 원본 데이터 평균
ebola['Cases_Liberia'].mean()
```

# 보간법의 다양한 종류
- method must be one of ['linear', 'time', 'index', 'values', 'nearest', 'zero', 'slinear', 'quadratic', 'cubic', 'barycentric', 'krogh', 'spline', 'polynomial', 'from_derivatives', 'piecewise_polynomial', 'pchip', 'akima', 'cubicspline']

# Linear 보간법
- 직선 거리에 따라 선형적으로 계산하는 방법

```python
ebola.interpolate(method = 'linear')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method = 'linear')['Cases_Liberia'].mean()
```

# Quadratic 보간법
- q(x) = ax^2 + bx + c
- 곡선으로 선을 그어 값을 대체하기 때문에 linear보다 그래프가 부드럽게 보정됨

```python
ebola.interpolate(method = 'quadratic')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method = 'quadratic')['Cases_Liberia'].mean()
```

# Zero 보간법
- 상수 그래프

```python
ebola.interpolate(method = 'zero')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method = 'zero')['Cases_Liberia'].mean()
```

# Cubic spline 보간법
- 3차 다항식으로 선 연결
- 앞에 연결된 곡선과 뒤에 연결된 곡선을 부드럽게 연결함

```python
ebola.interpolate(method = 'cubicspline')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method = 'cubicspline')['Cases_Liberia'].mean()
```

# Akima 보간법
- 로컬 근사화

```python
ebola.interpolate(method = 'akima')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method = 'akima')['Cases_Liberia'].mean()
```

# Pchip 보간법
- 형태 보존 조각별 3차 보간

```python
ebola.interpolate(method = 'pchip')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method = 'pchip')['Cases_Liberia'].mean()
```

```python
# 시각화를 하는 이유 - 패턴 찾기, 경향 파악, 분석 시작, 설명 보충
# 데이터분석가의 분석에 대한 근거를 제시하기 위해서 데이터분석에 대한 개념이 적은 사람들에게 시각화를 통해 쉽고 빠르게 내용 전달
# 시각화들을 통해 무엇을 알 수 있는지 명확히 알고 진행할 것
# 예: 산점도, 히스토그램, 막대그래프, 박스플랏
# 데이터 타입을 보고 어떤 그래프를 그릴지 결정
# 대표적인 시각화 라이브러리: matplotlib, seaborn
```

```python
import matplotlib.pyplot as plt
%matplotlib inline # .show() 함수 없이 바로 그래프 그려주는 매직커맨드
```

```python
# 산점도 그리기

plt.scatter([1,2,3],[4,5,6])
```

```python
# 산점도 여러개를 그리면 자동으로 합쳐짐

plt.scatter([1,2,3],[4,5,6])
plt.scatter([5,6,7],[10,5,6])
plt.scatter([2,4,8],[1,2,5])
```

```python
# 그래프 제목과 축 레이블 설정하기

plt.scatter([1,2,3],[4,5,6])
plt.title('mpg') # 그래프 제목 설정
plt.xlabel('mpg.hwy') # x축 레이블 설정
plt.ylabel('mpg.cty') # y축 레이블 설정
```

```python
# 축 범위 설정하기

plt.scatter([1,2,3],[4,5,6])
plt.xlim(0,200)
plt.ylim(30,50)
```

```python
# 축 눈금값 범례, 이름 설정

plt.scatter([1,2,3],[4,5,6])
plt.xticks([1,2,3], ['A','B','C'])
```

```python
# 그래프 범례(레이블) 출력하기

plt.scatter([1,2,3],[4,5,6], label = 'mpg.cty')
plt.scatter([6,7,8],[3,2,5], label = 'mpg.hwy')
plt.legend()
```

```python
# 그래프 범례(레이블) 위치 설정하기

plt.scatter([1,2,3],[4,5,6], label = 'mpg.cty')
plt.scatter([6,7,8],[3,2,5], label = 'mpg.hwy')
plt.legend(loc = 'center')
```

```python
# 그래프 크기 조절하기

plt.figure(figsize = (10,5))
plt.scatter([1,2,3],[4,5,6], label = 'mpg.cty')
```

```python
import seaborn as sns
```

```python
# 그리드 스타일 종류: white, dark, whitegrid, darkgrid, ticks

sns.set() # 기본값 그리드
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style = 'whitegrid') # 하얀색 그리드로 스타일 변경
plt.scatter([1,2,3],[4,5,6])
```

```python
sns.set(style = 'dark') # 어두운 그리드로 스타일 변경
plt.scatter([1,2,3],[4,5,6])
```

```python
# 선그래프 그려보기
# x, y 수치형 데이터

sns.set(style = 'whitegrid')
plt.plot([1,2,3],[4,5,6], label = 'mpg.cty')
plt.plot([6,7,8],[3,2,5], label = 'mpg.hwy')
plt.legend()
```

```python
# 선 색상 설정하기

plt.plot([1,2,3],[4,5,6], label = 'mpg.cty', color = 'red')
plt.plot([6,7,8],[3,2,5], label = 'mpg.hwy', color = 'blue')
plt.legend()
```

```python
# 선 스타일과 두께 설정하기

plt.plot([1,2,3],[4,5,6], label = 'mpg.cty', color = 'red', linestyle = 'dashed', linewidth = 10)
plt.plot([6,7,8],[3,2,5], label = 'mpg.hwy', color = 'blue', linestyle = 'dotted', linewidth = 5)
plt.legend()
```

```python
# 마커 설정하기

plt.plot([1,2,3],[4,5,6], marker = 's', label = 'mpg.cty', color = 'red', linestyle = 'dashed')
plt.plot([6,7,8],[3,2,5], marker = '*', label = 'mpg.hwy', color = 'blue', linestyle = 'dotted')
plt.legend()
```

```python
# 마커 사이즈 설정하기

plt.plot([1,2,3],[4,5,6], marker = 's', markersize = 10, label = 'mpg.cty', color = 'red', linestyle = 'dashed')
plt.plot([6,7,8],[3,2,5], marker = '*', label = 'mpg.hwy', color = 'blue', linestyle = 'dotted')
plt.legend()
```

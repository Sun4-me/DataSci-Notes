```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# mpg 데이터 2차원 시각화

```python
mpg = pd.read_csv('csv/mpg.csv')
mpg
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,8))

ax[0].scatter(mpg["displ"], mpg["hwy"], c='red', alpha=0.5)
ax[0].set_xlabel("displ")
ax[0].set_ylabel("hwy")
ax[0].set_title("displ vs. hwy")

counts = mpg["cyl"].value_counts()
ax[1].pie(counts, labels=counts.index, autopct="%1.1f%%", startangle=90)
ax[1].axis("equal")
ax[1].set_title("cyl")
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,8))

ax[0].scatter(mpg["displ"], mpg["cty"], alpha=0.5)
ax[0].set_xlabel("displ")
ax[0].set_ylabel("cty")
ax[0].set_title("displ vs. cty")

counts = mpg["drv"].value_counts()
ax[1].pie(counts, labels=counts.index, autopct="%1.1f%%", startangle=90)
ax[1].axis("equal")
ax[1].set_title("drv")
```

# titanic 데이터 2차원 시각화

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,8))

sns.scatterplot(x="age", y="fare", data=titanic, ax=ax[0])
ax[0].set_title("Age vs. Fare")

counts = titanic["sex"].value_counts()
ax[1].pie(counts, labels=counts.index, autopct="%1.1f%%", startangle=90)
ax[1].axis("equal")
ax[1].set_title("Gender")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,8))

sns.boxplot(x="class", y="age", data=titanic, ax=ax[0])
ax[0].set_title("Age by Class")

counts = titanic["embark_town"].value_counts()
ax[1].pie(counts, labels=counts.index, autopct="%1.1f%%", startangle=90)
ax[1].axis("equal")
ax[1].set_title("Embarked Town")
```

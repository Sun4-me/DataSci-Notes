# 다양한 마커들

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3],[4,5,6], marker = '.', color = 'red') # point
plt.plot([2,4,5],[4,5,3], marker = ',', color = 'green') # pixel
plt.plot([4,6,7],[3,5,7], marker = 'o', color = 'black') # circle
plt.plot([6,7,8],[3,2,5], marker = 'v', color = 'blue') # triangle_down
```

```python
plt.plot([1,2,3],[4,5,6], marker = '^', color = 'red') # triangle_up
plt.plot([2,4,5],[4,5,3], marker = '<', color = 'green') # triangle_left
plt.plot([4,6,7],[3,5,7], marker = '>', color = 'black') # triangle_right
plt.plot([6,7,8],[3,2,5], marker = '1', color = 'blue') # tri_down
```

```python
plt.plot([1,2,3],[4,5,6], marker = '2', color = 'red') # tri_up
plt.plot([2,4,5],[4,5,3], marker = '3', color = 'green') # tri_left
plt.plot([4,6,7],[3,5,7], marker = '4', color = 'black') # tri_right
plt.plot([6,7,8],[3,2,5], marker = '8', color = 'blue') # octagon
```

```python
plt.plot([1,2,3],[4,5,6], marker = 's', color = 'red') # square
plt.plot([2,4,5],[4,5,3], marker = 'p', color = 'green') # pentagon
plt.plot([4,6,7],[3,5,7], marker = 'P', color = 'black') # plus(filled)
plt.plot([6,7,8],[3,2,5], marker = '*', color = 'blue') # star
```

```python
plt.plot([1,2,3],[4,5,6], marker = 'h', color = 'red') # hexagon1
plt.plot([2,4,5],[4,5,3], marker = 'H', color = 'green') # hexagon2
plt.plot([4,6,7],[3,5,7], marker = '+', color = 'black') # plus
plt.plot([6,7,8],[3,2,5], marker = 'x', color = 'blue') # x
```

```python
plt.plot([1,2,3],[4,5,6], marker = 'X', color = 'red') # x(filled)
plt.plot([2,4,5],[4,5,3], marker = 'D', color = 'green') # diamond
plt.plot([4,6,7],[3,5,7], marker = 'd', color = 'black') # thin_diamond
plt.plot([6,7,8],[3,2,5], marker = '|', color = 'blue') # vline
plt.plot([1,2,3],[2,3,4], marker = '_', color = 'purple') # hline
```

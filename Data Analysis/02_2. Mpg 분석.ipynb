```python
import pandas as pd
import numpy as np
```

```python
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
# Q1 : 도시 연비 20 이상, 고속도로 연비 30 이하인 모델 추출

mpg.query('cty >= 20 & hwy <= 30').sort_values(['cty', 'hwy']).head()
    # 조건식 두가지
    # 오름차순 정렬 후 .head()로 추출

# impreza awd, forester awd, tiburon, orester awd, impreza awd
```

```python
# Q2 : 도시연비의 중앙값이 가장 높은 제조사는?

mpg.groupby('manufacturer').agg(cty_median = ('cty', 'median')).sort_values('cty_median', ascending = False)
    # 제조사별로 묶은 후 중간값을 내림차순으로 정렬

# honda
```

```python
# Q3: 고속도로 연비의 평균값이 가장 낮은 제조사는?

mpg.groupby('manufacturer').agg(hwy_mean = ('hwy', 'mean')).sort_values('hwy_mean', ascending = False)
    # 제조사별로 묶은 후 평균값을 내림차순으로 정렬
    
# honda
```

```python
# Q4: 파생변수를 사용하여 총 연비를 추가

mpg['tot'] = mpg['cty'] + mpg['hwy']
mpg
```

```python
# Q5: 총 연비의 평균값이 가장 높은 제조사는?
mpg.groupby('manufacturer').agg(tot_mean = ('tot', 'mean')).sort_values('tot_mean', ascending = False)
    # 제조사별로 묶은 후 내림차순으로 정렬
    
# honda
```

```python
# Q6 : 총연비 50 기준으로 전처리하기
mpg = mpg.assign(gb = np.where(mpg['tot'] > 50, 'good', 'bad'))
    # 총연비가 50보다크면 good, 작으면 bad
mpg
```

```python
# Q7 : 총 연비의 gb를 표현하고 좋은 모델의 수를 측정하시오.
mpg['gb'].value_counts().to_frame()

# 29개
```

```python
# Q8 : category 중 compact의 비율이 가장 높은 제조사는?
mpg=mpg.assign(category = np.where(mpg['category']== 'compact',1,0))
mpg.groupby('manufacturer').agg(compact = ('category','mean'))
# audi
```

```python
# Q9 : 제조사별 displ 평균은?
mpg.groupby('manufacturer').agg(mean_displ = ('displ','sum'))
```

```python
# Q10: 1999년에 생산된 honda 모델의 수는?

# 코드 작성
mpg_1 = mpg.query('manufacturer == "honda"')
    # query를 이용해 honda 제조사 데이터만 따로 저장
mpg_1['year'].value_counts().to_frame()
    # year를 기준으로 표 추출
    
# 5개
```

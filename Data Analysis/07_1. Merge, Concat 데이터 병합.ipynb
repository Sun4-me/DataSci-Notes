# 데이터 병합 Merge, Concat

```python
import pandas as pd
```

## Concat 함수
- 데이터프레임끼리 병합, 행 또는 열
- axis: 축
- ignore_index: 병합후 인덱스 새롭게 만들지에 대한 유무

- concat의 경우는 컬럼이 동일하다면 쉽게 붙일 수 있다.다만 시계열데이터는 다를 수 있으니 유의하기!

```python
df1 = pd.DataFrame({'A':[1,2,3,4], 'B':[1,2,3,4]})
df2 = pd.DataFrame({'A':[4,5], 'B':[4,5]})
df3 = pd.DataFrame({'A':[5,6], 'C':[3,2]})

display(df1)
display(df2)
display(df3)
```

```python
# 세로로 병합
pd.concat([df1,df3,df2]) # axis= 0 디폴트값
```

```python
 # 가로로 병합
pd.concat([df1,df3,df2], axis=1) # axis = 1
```

```python
pd.concat([df1,df2,df3], ignore_index = True)
```

```python
# 데이터프레임 나열 순서도 다르게 할 수 있다
pd.concat([df1,df3,df2], ignore_index=True)
```

## Merge 함수
- join key: 병합 기준

#### 함수 주요 인자

- left : 병합할 첫 번째 데이터 프레임으로 왼쪽 ( 순서가 기준 )
- right : 병합할 두 번째 데이터 프레임으로, 오른쪽 데이터 프레임 
- on : 조인 키로 사용할 컬럼 목록으로, 두 데이터프레임 모두 포함돼어야 한다.
- left_on : 왼쪽 데이터프레임에서 조인 키로 사용할 컬럼 목록
- right_on : 오른쪽 데이터프레임에서 조인키로 사용할 컬럼 목록
- left_index : 왼쪽 데이터프레임의 조인 키로 인덱스를 사용할지
- right_idnex: 오른쪽 데이터프레임의 조인키로 인덱스를 사용할지
- how : 조인의 방법을 결정하며, 디폴트는 'inner'

```python
df1 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '영어':[100,50,60,70]})
df2 = pd.DataFrame({'이름':['홍길동','김영희','김철수'],
                   '수학':[50,100,60]})

display(df1)
display(df2)
```

```python
pd.merge(df1,df2) # 디폴트: inner (교집합)
```

```python
df3 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '수학':[50,100,60,70],
                   '생년월일':['00','01','02','03']})
df4 = pd.DataFrame({'이름':['홍길동','김영희','김철수'],
                   '수학':[50,100,60],
                   '지각횟수':[0,1,2]})

display(df3)
display(df4)
```

```python
pd.merge(df3,df4, on=['이름'])
```

```python
# on을 사용할 때 한 개의 컬럼만 사용하는 게 아니라 2개 이상 컬럼도 사용할 수 있다.
# 이 경우, 수학이라는 컬럼도 공통되기 때문에 이름과 수학 두개의 컬럼 사용하는 것이 좋다.
pd.merge(df3,df4, on=['이름','수학'])
```

```python
df3 = pd.DataFrame({'이름':['홍길동','김영희','김철수','임꺽정'],
                   '수학':[50,100,60,70],
                   '생년월일':['00','01','02','03']})
df4 = pd.DataFrame({'name':['홍길동','김영희','김철수'],
                   '수학':[50,100,60],
                   '지각횟수':[0,1,2]})

display(df3)
display(df4)
```

```python
pd.merge(df3,df4, left_on=['이름'],right_on=['name'])
```

```python
pd.merge(df3,df4, left_on=['이름','수학'],right_on=['name','수학'])
```

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[1,2,3]})
df2 = pd.DataFrame({'A':[2,3,4], 'C':[7,8,9]})

display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on='A')
```

```python
# 왼쪽과 오른쪽 데이터프레임 조인키에 대한 교집합 기준 
pd.merge(df1, df2, on='A',how = 'inner')
```

```python
# 왼쪽 데이터프레임 조인 키를 기준 병합
pd.merge(df1, df2, on='A',how = 'left')
```

```python
# 오른쪽 데이터프레임의 조인 키를 기준 병합
pd.merge(df1, df2, on='A',how = 'right')
```

```python
# 왼쪽과 오른쪽 데이터프레임 조인키에 대한 합집합 기준 
pd.merge(df1, df2, on='A',how = 'outer')
```

# 범례 값 만들어서 비교해 보기
- left
- center
- right
- upper left , right, center
- lower left, right, center

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'center') # 중앙에 배치
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'center left') # 중앙 왼쪽에 배치
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'center right') # 중앙 오른쪽에 배치
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'upper center') # 위 중앙에 배치
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'upper left') # 위 왼쪽에 배치
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'upper right') # 위 오른쪽에 배치
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'lower center') # 아래 중앙에 배치
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'lower left') # 아래 왼쪽에 배치
```

```python
plt.scatter([1,2,3],[4,5,6], label = 'A')
plt.scatter([6,7,8],[3,2,5], label = 'B')
plt.legend(loc = 'lower right') # 아래 오른쪽에 배치
```

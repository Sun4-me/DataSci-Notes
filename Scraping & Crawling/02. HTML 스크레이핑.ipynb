```python
html_str = """
<html>
  <head>
    <title>안녕하세요</title>
  </head>
  <body>
    <div id="container">
      <p class='p1'>hello</p>
      <p>Bye</p>
    </div>
  </body>
</html>"""
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_str, 'html.parser')
soup
```

```python
type(soup)
```

```python
container_div = soup.find("div", {"id":"container"})
container_div
```

```python
type(container_div)
```

```python
soup.find_all("p")
```

```python
# 클래스가 p1인 p 엘리먼트 모두 찾아보기
soup.find_all("p", {'class': 'p1'})

# all은 리스트로 추출됨
```

```python
# 텍스트 추출하려면 element 객체인 상태에서만 가능
soup.find('p', {'class':'p1'}).text
```

```python
# 두 개의 p 태그 내에 있는 텍스트 추출하기
p_tags = soup.find_all("p")
print(p_tags[0].text)
print(p_tags[1].text)
```

```python
for p in p_tags:
    print(p.text)
```

```python
GMARKET_BEST_PRODUCT_URL = "https://corners.gmarket.co.kr/BestSellers"
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get(GMARKET_BEST_PRODUCT_URL)
res
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
soup
```

```python
best_item_sample = soup.select_one("#gBestWrap > div.gbest-top > div > div.best-list > ul > li:nth-child(23)")
best_item_sample
```

```python
best_item_sample.select_one("a.itemname").text
```

```python
best_item_sample.select_one('img.lazy')
```

```python
image_url = "https:"+ best_item_sample.select_one('img.lazy')['data-original']
print(image_url)
```

```python
from urllib.request import urlretrieve
urlretrieve(image_url, "gmarket_image.png")
```
